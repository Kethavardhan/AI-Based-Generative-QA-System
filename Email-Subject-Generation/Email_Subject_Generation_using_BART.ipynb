```python
def tokenize_inputs(email_content):
  #print('tokenize_inputs')
  #print(type(email_content))
  start_prompt = "Generate subject for below email: \n\n"
  end_prompt = "\n\nSubject: "
  prompts = [start_prompt + dialogue + end_prompt for dialogue in email_content['input_text']]

  model_inputs = tokenizer(prompts, text_target=email_content['target_text'],max_length=128, truncation=True, padding='max_length', return_tensors='pt')
  #print(model_inputs.keys())
  return model_inputs
```

```python
tokenizer.pad_token = tokenizer.eos_token
tokenized_datasets_train = hf_train_dataset.map(tokenize_inputs, batched=True)
tokenized_datasets_train = tokenized_datasets_train.remove_columns(['input_text', 'target_text'])
tokenized_datasets_train = tokenized_datasets_train.filter(lambda example, index: index % 100 == 0, with_indices=True) # Shortening of data, could use: shuffle and selec
```

```python
def generate_prompt(email_content, llm, tokenizer,device="cuda"):
    #print('generate_prompt new method called')
    # Move the model to the specified device
    llm = llm.to(device)

    #Send a clear and concise subject line for the following email.

    # Construct the input prompt
    input_prompt = f"""
                    Generate subject for below email:

                    {email_content}

                    Subject:
                    """

    # Tokenize the input prompt
    input_ids = tokenizer(input_prompt, return_tensors='pt', max_length=512)

    # Move the input tensor to the specified device
    input_ids = {key: value.to(device) for key, value in input_ids.items()}

    # Generate output on the same device
    tokenized_output = llm.generate(input_ids['input_ids'], min_length=10, max_length=128)

    # Move the output tensor back to CPU if needed
    tokenized_output = tokenized_output.to("cpu")

    # Decode the generated output
    output = tokenizer.decode(tokenized_output[0], skip_special_tokens=True)

    return output

sample_email = """Phillip,   Could you please do me a favor?
I would like  to read your current title policy to see what it says about easements.
You  should have received a copy during your closing.
I don't know how many  pages it will be but let me know how you want to handle getting a copy  made.
I'll be happy to make the copy, or whatever makes it easy for  you.
Thanks,"""

label = 'Huntley/question'

output = generate_prompt(sample_email, model, tokenizer)
print("Sample")
print(sample_email)
print("-------------------")
print("Model Generated Summary:")
print(output)
print("Correct Summary:")
print(label)
```

```python
from evaluate import load
from collections import defaultdict
from tqdm import tqdm

def metric(data):
    sub,gen_subject  = ["subject","gen_subject"]
    rouge = load("rouge")
    meteor = load("meteor")
    bleu = load("sacrebleu")
    score = defaultdict(int)

    for _,x in tqdm(data.iterrows(),total=len(data)):

        rouge_s = rouge.compute(predictions=[x[gen_subject]],references=[x[sub]])

        score["rouge_1_sub"] += rouge_s['rouge1']
        score["rouge_2_sub"] += rouge_s['rouge2']
        score["rouge_L_sub"] += rouge_s['rougeL']

        meteor_s = meteor.compute(predictions=[x[gen_subject]],references=[x[sub]])

        score["meteor_sub"] += meteor_s["meteor"]

        bleu_s = bleu.compute(predictions=[x[gen_subject]],references=[x[sub]])

        score["bleu_s"] += bleu_s["score"]

    print("The evalution scores are as follows:")

    for name,val in score.items():
        print(f"{name}: {val/len(data)}")
```

```python
import gradio as gr

tokenizer_hfhub = AutoTokenizer.from_pretrained("Nishantc05/emailSubGen-bartmodel")
model_hfhub = AutoModelForSeq2SeqLM.from_pretrained("Nishantc05/emailSubGen-bartmodel")
model_hfhub.to(device)
```

```python
def generate_subject(email_content):

    # Construct the input prompt
    input_prompt = f"""
                    Generate subject for below email:

                    {email_content}

                    Subject:
                    """

    # Tokenize the input prompt
    input_ids = tokenizer_hfhub(input_prompt, return_tensors='pt', max_length=512)

    # Move the input tensor to the specified device
    input_ids = {key: value.to(device) for key, value in input_ids.items()}

    # Generate output on the same device
    tokenized_output = model_hfhub.generate(input_ids['input_ids'], min_length=10, max_length=128)

    # Move the output tensor back to CPU if needed
    tokenized_output = tokenized_output.to("cpu")

    # Decode the generated output
    output = tokenizer_hfhub.decode(tokenized_output[0], skip_special_tokens=True)

    return output
```
